## 1. plant data 가져오기

```
# 모든 날짜별 + 기상 + trend + point 데이터
data = pd.read_csv(os.path.join(DATA_CONV_PATH, "data_all_time_predict.csv"))

data.drop(['Unnamed: 0'], axis='columns', inplace=True)

data
```

```
data.info()
```

# Modeling


```
# out을 예측하는 모델 가져오기
model_tem_out_loc = joblib.load(os.path.join(MODEL_PATH, "model_tem_out_loc.pkl"))
model_hum_out_loc = joblib.load(os.path.join(MODEL_PATH, "model_hum_out_loc.pkl"))

model_out_loc = [model_tem_out_loc, model_hum_out_loc]
model_out_loc_name = ['tem_out_loc', 'hum_out_loc']
```

```
# 예측에 필요한 변수 정의
data_columns = list(data.columns)
remove_list = ['date']
for rmv in remove_list:
  data_columns.remove(rmv)
data_columns
```

```
# out 예측하기
for i in range(len(model_out_loc_name)):
  data[model_out_loc_name[i]] = model_out_loc[i].predict(data[data_columns])
data
```

```
data.to_csv(os.path.join(DATA_CONV_PATH, "data_all_time_out.csv"))
```
