## 2. Check NULL

```
def countNull(df):
  """
  describe : 데이터 갯수, NULL 갯수, 백분율을 보기위한 코드
  <input>
  pd.dataframe
  <output>
  pd.dataframe
  """
  count_null = pd.DataFrame()
  count_null['numOfTotal'] = [len(df) for x in df.columns]
  count_null['numOfNull'] = list(df.isnull().sum())
  count_null['numOfNotnull'] = count_null['numOfTotal'] - count_null['numOfNull']
  count_null['percent'] = count_null['numOfNull']/count_null['numOfTotal']*100
  count_null['dataType'] = [type(df[x][0]) for x in df.columns]
  count_null.index = df.columns
  return count_null
```

## 1. 사용하는 함수

```
def arrangeData(df, plant):
  """
  describe : plant, loc에 따라 column을 볼 수 있게 데이터 프레임을 만든다.
  """
  ret_columns_name = ['mea_ddhr','plant','loc','tem_in','hum_in','tem_coil','tem_out_loc','hum_out_loc','cond_loc']
  ret_df = pd.DataFrame(columns=ret_columns_name)
  # plant = int(df.columns[1].split('_')[0][-1])
  np_df = np.array(df)
  ret_df = np.repeat(' '*30, len(df)*3*9).reshape(len(df)*3, 9)
  for index in range(len(df)):
    for loc in range(3):
      p = 3*loc + 1
      ret_df[3*index + loc,0] = np_df[index, 0]
      ret_df[3*index + loc,1] = plant
      ret_df[3*index + loc,2] = loc + 1
      ret_df[3*index + loc,3] = np_df[index, p] # tem_in
      ret_df[3*index + loc,4] = np_df[index, p+1] # hum_in
      ret_df[3*index + loc,5] = np_df[index, p+2] # tem_coil
      ret_df[3*index + loc,6] = np_df[index, 10] # tem_out
      ret_df[3*index + loc,7] = np_df[index, 11] # hum_out
      ret_df[3*index + loc,8] = np_df[index, 12+loc]
  ret_df = pd.DataFrame(ret_df, columns=ret_columns_name)
  for col in ret_columns_name:
    if col == 'mea_ddhr':
      ret_df[col] = ret_df[col].apply(lambda x : str(x) if x != 'nan' else np.nan)
    else:
      ret_df[col] = ret_df[col].apply(lambda x : np.float64(x) if x != 'nan' else np.nan)
  return ret_df
```

```
import datetime
def stringToDatetime(df, colName='mea_ddhr'):
  """
  describe : string 을 datetime으로 변경한다.
  input = pd.Dataframe
  output = pd.Dataframe
  """
  if type(df[colName][0]) != str: # 이미 자료형이 datetime인 경우 return 한다.
    return df
  if len(df[colName][0].split(' ')[1].split(':')) == 2: # 날짜 정보에 초에 대한 정보가 없을 경우 초를 붙여주고 datetime으로 변경한다.
    df[colName] = df[colName].apply(lambda x : x + ':00' if x != 'nan' else np.nan)
  df[colName] = df[colName].apply(lambda x : datetime.datetime.strptime(x, '%Y-%m-%d %H:%M:%S') if x != 'nan' else np.nan)
  return df

def createDateTime(df, colName='mea_ddhr'):
  """
  describe : datetime의 정보에서 date 정보와 0시 기준 몇 분인지를 나타내는 time을 분리하여 추가한다.
  input = pd.Dataframe
  output = pd.Dataframe
  """
  df = stringToDatetime(df, colName)
  df['date'] = df[colName].apply(lambda x : x.date())
  df['time'] = df[colName].apply(lambda x : np.float64(x.hour * 60 + x.minute))
  return df
```
