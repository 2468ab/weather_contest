# IN, OUT 클러스터링

## 1. Trend clustering

```
# trend, point 클러스터링을 하기 위한 변수 이름
tem_name = ['tem_out_loc', 'hum_out_loc', 'tem_in', 'hum_in', 'tem_coil']
```

```
# trend 함수
import pickle
from sklearn.externals import joblib
def trend_fn(df, plant, loc, target, k=5):
  temp_trend = pd.DataFrame(columns=['date','0-4', '4-8','8-12','12-16','16-20','20-24'])

  date_start = datetime.strptime("2016-04-01 00:00:00", '%Y-%m-%d %H:%M:%S').date()
  date_end = datetime.strptime("2020-03-31 00:00:00", '%Y-%m-%d %H:%M:%S').date()
  date = date_start

  while date <= date_end:
    if date.month == 1 and date.day == 1:
      print(plant, loc, target, date)
    temp_date = pd.DataFrame(columns=['date','0-4', '4-8','8-12','12-16','16-20','20-24'])
    temp_date.loc[0] = 0
    temp_date['date'] = date
    for time in range(0, 24, 4):
      start_time = time * 60
      end_time = (time + 4) * 60
      if end_time == 24 * 60:
        end_time = 0
      start = df[(df['date'] == date) & (df['time'] == start_time)][target]
      end = df[(df['date'] == date) & (df['time'] == end_time)][target]
      temp_date['{}-{}'.format(time, time + 4)] = start[start.index[0]] - end[end.index[0]]
    temp_trend = temp_trend.append(temp_date)  # append를 사용하면 concat이랑 같은 효과임
    date += timedelta(days=1)
  #
  # 각 일에 대한 interpolation 함수를 대변하는 feature를 기반으로 클러스터링을 진행한다.
  kmeans = KMeans(n_clusters=k)
  kmeans.fit(temp_trend[['0-4', '4-8','8-12','12-16','16-20','20-24']])
  temp_trend[target + '-trend'] = kmeans.predict(temp_trend[['0-4', '4-8','8-12','12-16','16-20','20-24']])
  temp_trend['plant'] = plant
  temp_trend['loc'] = loc
  #print(temp_trend[['date', 'plant', 'loc', target + '-trend']])
  return temp_trend[['date', 'plant', 'loc', target + '-trend']]
```

```
# trend 만들기
for tem in tem_name:
  cluster_df[tem + '-trend'] = -1
for tem in tem_name:
  data_tem = data[['date', 'plant', 'loc', 'time', tem]]
  for plant in range(1, 3):
    for loc in range(1, 4):
      data_slice = data_tem[(data_tem['plant'] == plant) & (data_tem['loc'] == loc)]
      data_slice = trend_fn(data_slice, plant, loc, tem)
      temp_cluster_df = cluster_df[(cluster_df['plant'] == plant) & (cluster_df['loc'] == loc)]
      temp_cluster_df[tem + '-trend'] =  list(data_slice[tem + '-trend'])
      cluster_df.loc[temp_cluster_df.index] = temp_cluster_df
cluster_df
```

## 2. Point Clustering

```
# point 함수
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

def point_fn(df, plant, loc, target, k=5):
  timeList = list(range(24))
  timeList.insert(0, "min")
  timeList.insert(0, "max")
  timeList.insert(0, "date")
  timeList = list(map(str, timeList))

  temp_point = pd.DataFrame(columns=timeList)

  date_start = datetime.strptime("2016-04-01 00:00:00", '%Y-%m-%d %H:%M:%S').date()
  date_end = datetime.strptime("2020-03-31 00:00:00", '%Y-%m-%d %H:%M:%S').date()
  date = date_start

  while date <= date_end:
    tem_df = df[df['date'] == date]
    if date.month == 1 and date.day == 1:
      print(plant, loc, target)
    
    temp_date = pd.DataFrame(columns=timeList)
    temp_date.loc[0] = 0
    temp_date['date'] = date
    # Max, Min column 넣어주기
    temp_date["min"] = np.min(list(tem_df[target]))
    temp_date["max"] = np.max(list(tem_df[target]))

    # 시간대에 맞는 값 넣어주기
    for hour in range(24):
      if hour == 24:
        hour = 0
      time = hour * 60
      value = list(tem_df[tem_df['time'] == time][tem])[0]
      temp_date[f'{hour}'] = value
    
    # 만들어진 df 합치기
    temp_point = temp_point.append(temp_date)
    date += timedelta(days=1)

  # 만들어진 temp_point 를 가지고, PCA를 수행한다.
  x = temp_point.drop(['date', "min", "max"], axis=1).values # 독립변인들의 value값만 추출
  x = StandardScaler().fit_transform(x) # x객체에 x를 표준화한 데이터를 저장
  
  pca = PCA(n_components=24) # 주성분을 몇개로 할지 결정
  printcipalComponents = pca.fit_transform(x)

  # eigen value를 통해, 몇개의 축을 선정할 것인지 판단하자.
  print('explained variance ratio :', pca.explained_variance_ratio_)

  # 주성분으로 이루어진 데이터 프레임 구성
  pca_df = pd.DataFrame(data= printcipalComponents[:,:2], columns=["pc1", "pc2"])

  # index 초기화
  temp_point = temp_point.reset_index()
  pca_df = pca_df.reset_index()

  # df 합치기
  temp_point = pd.merge(temp_point, pca_df, left_index=True, right_index=True, how='left')
  temp_point = temp_point.drop(["index_x", "index_y"], axis=1)
  # print(temp_point)

  # clustering을 위한 df
  cluster_df = temp_point[["pc1", "pc2", "min", "max"]]
  # print(cluster_df)

  kmeans = KMeans(n_clusters=k)
  kmeans.fit(cluster_df)
  temp_point[target+'-point'] = kmeans.predict(cluster_df)
  temp_point['plant'] = plant
  temp_point['loc'] = loc
  #print(temp_point[['date', 'plant', 'loc', target + '-point']])
  return temp_point[['date', 'plant', 'loc', target + '-point']]
```

```
# point 만들기
for tem in tem_name:
  cluster_df[tem + '-point'] = -1
for tem in tem_name:
  data_tem = data[['date', 'plant', 'loc', 'time', tem]]
  for plant in range(1, 3):
    for loc in range(1, 4):
      data_slice = data_tem[(data_tem['plant'] == plant) & (data_tem['loc'] == loc)]
      data_slice = point_fn(data_slice, plant, loc, tem)
      temp_cluster_df = cluster_df[(cluster_df['plant'] == plant) & (cluster_df['loc'] == loc)]
      temp_cluster_df[tem + '-point'] =  list(data_slice[tem + '-point'])
      cluster_df.loc[temp_cluster_df.index] = temp_cluster_df
cluster_df
```
